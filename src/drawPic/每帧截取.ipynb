```python
# wavLenth = len(signal)
wavLenth = 44100
frameTimeSize = 25
moveTimeSize = 10
# 25ms 截取一帧 那么帧长取决于采样率 44100次是1秒 那么 44100 * 25 /1000 =  1102.5
# frameSize = math.ceil(sampleRate * frameTimeSize/ 1000)  # 帧中包含多少次采样数
# moveSize = math.ceil(sampleRate * moveTimeSize/ 1000)  # 每次移动多少次采样数
frameSize = sampleRate * frameTimeSize/ 1000 # 帧中包含多少次采样数
moveSize = sampleRate * moveTimeSize/ 1000  # 每次移动多少次采样数
totalMovmentTimes =  (wavLenth -  frameSize)  /  moveSize # 总共的移动次数
print("每一帧采样数为{}".format(frameSize))
print("窗口每次移动为{}".format(moveSize))
print("窗口可以移动{}次".format(totalMovmentTimes)) 
```
